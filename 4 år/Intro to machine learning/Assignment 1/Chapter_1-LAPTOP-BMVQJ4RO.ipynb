```python
import matplotlib.pyplot as plt
import numpy as np

# Train the model
log_reg = LogisticRegression(learning_rate=0.1, epochs=1000)
log_reg.fit(X_train, y_train)

# Visualize the loss and accuracy over epochs
epoch_list = np.arange(0, log_reg.epochs, 1)

fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting loss
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss', color='red')
loss_line, = ax1.plot(epoch_list, log_reg.losses, color='red', label='Loss')
print(log_reg.train_accuracies[-1])
ax1.tick_params(axis='y', labelcolor='red')

# Create a second y-axis for accuracy
ax2 = ax1.twinx()
ax2.set_ylabel('Accuracy', color='blue')
accuracy_line, = ax2.plot(epoch_list, log_reg.train_accuracies, color='blue', label='Accuracy')
ax2.tick_params(axis='y', labelcolor='blue')

# Combine legends from both axes
lines = [loss_line, accuracy_line]
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc="center right")

# Title and layout
fig.suptitle('Loss and Accuracy Over Time')
fig.tight_layout()  # Adjusts the plot to ensure everything fits without overlapping
plt.show()


# Normalize the losses and accuracies to bring them to a similar scale (plot 2)
max_loss = max(log_reg.losses)
min_loss = min(log_reg.losses)
normalized_losses = [(loss - min_loss) / (max_loss - min_loss) for loss in log_reg.losses]

max_acc = max(log_reg.train_accuracies)
min_acc = min(log_reg.train_accuracies)
normalized_accuracies = [(acc - min_acc) / (max_acc - min_acc) for acc in log_reg.train_accuracies]

# Visualize the normalized loss and accuracy over epochs
epoch_list = np.arange(0, log_reg.epochs, 1)

plt.figure(figsize=(10, 6))
plt.plot(epoch_list, normalized_losses, color='red', label='Loss')
plt.plot(epoch_list, normalized_accuracies, color='blue', label='Accuracy')

plt.xlabel('Epoch')
plt.ylabel('Normalized Value')
plt.title('Loss and Accuracy Over Time (normalized values)')
plt.legend()
plt.show()

print("A) as shown with the first graph above the loss starts at", log_reg.losses[0]," and ends at ", log_reg.losses[-1], " for the training data")
print("On the other hand, the accuracy of the predictions starts at ", log_reg.train_accuracies[0], " and ends on ", log_reg.train_accuracies[-1], " for the training data")
print(". But it is easily noticeable that the tradeoff between epochs/iterations and pay off (more accuracy and less loss) quickly diminishes after about 100 iterations")
print(" It is also essential to note that using only linear combinations of the inputs gave bad results, to address this i switched to quadratic features instead")
print("This was done by introducing new input terms, X_0^2, X_1^2 and the interaction term X_0 * X_1")

print("")
print("")

print("B) Inductive bias is the assumptions our model makes for generalizing from training data to unseen/test data. This is important because it dictitates how a model" 
      , " generalises to new unseen data.")
print("In this case the first logistic regression model had an inductive bias of assuming there was a linear relation ship between the inputs x_0, x_1 and the result y ")
print("This made the model terrible and this inductive bias had to be modified, in this case to quadratic")
```

```python
from sklearn.metrics import mean_squared_log_error

def rmsle(y_true, y_pred):
    """ Root Mean Squared Logarithmic Error """
    return np.sqrt(mean_squared_log_error(y_true, y_pred))
```
