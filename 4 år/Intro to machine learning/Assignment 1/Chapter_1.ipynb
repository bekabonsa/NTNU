```python
import matplotlib.pyplot as plt


# Load the dataset
train = pd.read_csv('mission3_train.csv')
test = pd.read_csv('mission3_test.csv')

# Create a new feature to check if data_stream_3 last digit is odd or even
train['data_stream_3_even_odd'] = (train['data_stream_3'] * 1000).astype(int) % 2
test['data_stream_3_even_odd'] = (test['data_stream_3'] * 1000).astype(int) % 2

# Include this new feature along with the existing features
X_train_new = train.drop(columns=['target']).values
y_train = train['target'].values

X_test_new = test.drop(columns=['target']).values
y_test = test['target'].values

# Train the Decision Tree with the new feature
clf_new = DecisionTreeClassifier(max_depth=6, criterion="log_loss", random_state=42)
clf_new.fit(X_train_new, y_train)

# Make predictions and evaluate the new model
y_pred_new = clf_new.predict(X_test_new)
y_proba_new = clf_new.predict_proba(X_test_new)[:, 1]

# Calculate ROC AUC score
auc_new = roc_auc_score(y_test, y_proba_new)
print(f"ROC AUC with new feature: {auc_new:.3f}")

```

 a. Explain your threshold-breaking strategy. Did you change the default hyperparameters?

 Out of the countless attempts of fine tuning the model and careful sampling of data, the only significant increase in ROC AUC came when checking if data stream 3 * 1000 was odd or not and adding that to the list of important features. This was deduced from the hint given out on blackboard. Super secure encryption is generating a random value that is always even.
 So depending on wethere the message m is odd or even the output c is odd or even. If m is odd m+random is also odd, if m is even m+random is also even. This becomes a good attack vector as we can now exploit it by making use of the extra bit of information for our model.

Max depth which is a hyperparameter was also changed to 6 to get a roc auc score of 0.72

```python
from sklearn.metrics import roc_curve, roc_auc_score, classification_report
import matplotlib.pyplot as plt
import numpy as np

# Calculate FPR, TPR, and thresholds
fpr, tpr, thresholds = roc_curve(y_test, y_proba_new)

# Calculate Youden's J statistic (TPR - FPR) to find the best threshold
j_scores = tpr - fpr  
best_threshold = thresholds[np.argmax(j_scores)]

print(f"Best threshold: {best_threshold}")

# Plot the ROC curve
plt.plot(fpr, tpr, label='ROC curve (AUC={:.3f})'.format(roc_auc_score(y_test, y_proba_new)))
plt.plot([0, 1], [0, 1], color='black', linestyle='--', label='Random (AUC=0.5)')

# Highlight the best threshold point on the ROC curve
plt.scatter(fpr[np.argmax(j_scores)], tpr[np.argmax(j_scores)], color='red', label=f'Sweet spot: {best_threshold:.2f}', zorder=5)

plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve with sweet spot Highlighted')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

# Print classification report for decision tree predictions at the best threshold
y_pred_custom_threshold = (y_proba_new >= best_threshold).astype(int)
#print(y_pred_custom_threshold)
print("\nClassification Report for decision tree:")
print(classification_report(y_test, clf_new.predict(X_test_new)))
#print(classification_report(y_test, y_pred_custom_threshold))
```

 b. Justify ROC AUC usage. Plot and interpret ROC.

 A score of 0.5 for ROC AUC indicates random guessing and is like a coin flip for distinguishing between the two classes. A score of 1 indicates a perfect model which gets it right  100% of the time. 
 An ROC-curve shows the relationship between TPR and FPR accross every threshold. An ROC-curve is a good choice for our case because the data is imbalanced (It shows the relation of both the positive class and the negative class). 
  
If every misclassifcation risks losing a sector of the city to machine control, we would want to minimize FPR and FNR while maximizing TPR and TNR. For our case it would be important to classify all critical communications as critical so maximizing TPR while decreasing FPR from our ROC curve is a good step forward. A good threshold for this can be a spot close to the point (0.5, 0.85) with a threshold of 0.21 (probability values greater than 0.21 will be considered a positive class). But if we look at the classification report the precision for class 1 is 0.56 which means it has a lot of false positives (which we would like to avoid).

e. In real life, you do not have the test targets. How would you make sure your model will work good on the unseen data? 

The way to do this is to use cross-validation. A process where you split the training data into smaller sets (folds) for training and testing multiple times while shuffling the folds and test.
As an example, you can create K number of folds and train it k times, with each of the training sessions having a different fold as the validation set. If you have 5 folds, train 5 times, with the validation set being changed 5 times to a different fold.

In addition, you want to check for overfitting by seeing the performance on the validation set. If it is terrible on the validation set it means you have a model that is overfitting and generalizes badly to new unseen data.
